### Step 1: Import Libraries

```python
import random
import pandas as pd
```

### Step 2: Simulate dice rolls

```python
# State Variables
n_rolls = 1000
results = []

# Dice roll simulation

for _ in range (n_rolls):
    r = random.random() # random float
    if r < 1/6:
        results.append(1)
    elif r < 2/6:
        results.append(2)
    elif r < 3/6:
        results.append(3)
    elif r < 4/6:
        results.append(4)
    elif r < 5/6:
        results.append(5)
    else:
        results.append(6)
```

### Step 3: Count frequencies

```python
frequency = [results.count(i) for i in range(1,7)]
```

### Step 4: Compute probabilities

```python
probability = [f / n_rolls * 100 for f in frequency]
```

### Step 5: Create and display a table

```python
dice_table = pd.DataFrame({
    'Die Face': list(range(1,7)) + ['Total'],  
    'Frequency': frequency + [sum(frequency)],        
    'Probability (%)': probability + [sum(probability)]  
})
dice_table.style.hide(axis='index').format({'Probability (%)': '{:.2f}'})
```
